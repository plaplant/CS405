```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc("text", usetex=True)
matplotlib.rc("font", family="serif")
matplotlib.style.use("dark_background")
```

```python
def square_wave(x):
    nwraps = x // (np.pi / 2)
    return np.where(nwraps % 2 == 0, 1, -1)
```

```python
def approx(x):
    return (
        4 * np.sin(2 * x) / np.pi
        + 4 * np.sin(6 * x) / (3 * np.pi)
        + 4 * np.sin(10 * x) / (5 * np.pi)
        + 4 * np.sin(14 * x) / (7 * np.pi)
    )
```

```python
xvals = np.linspace(-5, 5, num=1000)
yvals = square_wave(xvals)
yapprox = approx(xvals)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, yvals)
ax.plot(xvals, yapprox, alpha=0.7)
ax.set_xlabel("$t$", fontsize=20)
ax.set_ylabel("$f(t)$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("fourier0.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.linspace(-5, 5, num=1000)
yvals = square_wave(xvals)
y0 = 4 * np.sin(2 * xvals) / np.pi
y1 = 4 * np.sin(6 * xvals) / (3 * np.pi)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, yvals)
ax.plot(xvals, y0, alpha=0.7, label="First term")
ax.plot(xvals, y1, alpha=0.7, label="Second term")
ax.set_xlabel("$t$", fontsize=20)
ax.set_ylabel("$f(t)$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)
ax.legend(loc=1, fontsize=14)

fig.savefig("fourier1.pdf", bbox_inches="tight", transparent=True)
```
