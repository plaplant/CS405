```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc("text", usetex=True)
matplotlib.rc("font", family="serif")
```

```python
# generate random numbers from uniform distribution
rvs = stats.uniform.rvs(0, 1, size=(100,))

# compute mean and std
mean = np.mean(rvs)
std = np.std(rvs)

# make plot
fig = plt.figure()
ax = plt.gca()
hist, bins, _ = ax.hist(rvs, bins=10)

xvals = np.linspace(-3, 3, num=1000)
yvals = stats.norm.pdf(bins, mean, std)
ax.plot(bins, yvals / yvals.sum() * len(rvs))
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("$p(x)$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)
```

```python
# perform KS test
stats.kstest(rvs, stats.norm.cdf)
```

```python
# let's do the same with a normal distribution
# generate random numbers from uniform distribution
rvs = stats.norm.rvs(0, 1, size=(100,))

# compute mean and std
mean = np.mean(rvs)
std = np.std(rvs)

# make plot
fig = plt.figure()
ax = plt.gca()
hist, bins, _ = ax.hist(rvs, bins=10)

xvals = np.linspace(-3, 3, num=1000)
yvals = stats.norm.pdf(bins, mean, std)
ax.plot(bins, yvals / yvals.sum() * len(rvs))
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("$p(x)$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)
```

```python
# perform KS test
stats.kstest(rvs, stats.norm.cdf)
```
