# Matplotlib

## Basics

```python
# let's make some data to plot
xvals = np.linspace(0, 2 * np.pi, num=1000)
yvals = np.sin(xvals)
```

```python
# make a plot!
plt.figure()
plt.plot(xvals, yvals)
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# now let's make multiple lines and add a legend + title
y1 = np.sin(xvals)
y2 = np.cos(xvals)
plt.figure()
plt.plot(xvals, y1, label=r"$\sin x$", linestyle="-")
plt.plot(xvals, y2, label=r"$\cos x$", linestyle="--")
plt.legend()
plt.title("Trig is fun!")
plt.show()
```

## Bar Chart

```python
# bar chart
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

## Pie Chart

```python
# pie chart
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

## Pseudo-color plot

```python
# generate data
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
# matplotlib supports 3d plots
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(
    X,
    Y,
    Z,
    cmap="coolwarm",
    linewidth=0,
    antialiased=False,
)
```

```python
# but! much more common to see 2d pseudo-color plots
fig, ax = plt.subplots()
im = ax.imshow(
    Z,
    interpolation='bilinear',
    cmap="coolwarm",
    origin='lower',
    extent=[-3, 3, -3, 3],
    vmax=abs(Z).max(),
    vmin=-abs(Z).max(),
)
plt.colorbar(im)
plt.show()
```
