```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import stats
```

```python
matplotlib.rc("text", usetex=True)
matplotlib.rc("font", family="serif")
plt.style.use("dark_background")
```

```python
xvals = np.linspace(-4, 4, num=1000)
yvals = stats.norm.pdf(xvals, 0, 1)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, yvals)
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("PDF($x$)", fontsize=20)
ax.set_title("$\mu = 0,\, \sigma = 1$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("gaussian_pdf.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.linspace(-4, 4, num=1000)
yvals = stats.norm.cdf(xvals, 0, 1)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, yvals)
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("CDF($x$)", fontsize=20)
ax.set_title("$\mu = 0,\, \sigma = 1$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("gaussian_cdf.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.linspace(-4, 4, num=1000)
y_0_1 = stats.norm.pdf(xvals, 0, 1)
y_0_2 = stats.norm.pdf(xvals, 0, 2)
y_0_half = stats.norm.pdf(xvals, 0, 0.5)
y_2_1 = stats.norm.pdf(xvals, 2, 1)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, y_0_1, label=r"$\mu=0, \sigma=1$")
ax.plot(xvals, y_0_2, label=r"$\mu=0, \sigma=2$")
ax.plot(xvals, y_0_half, label=r"$\mu=0, \sigma=1/2$")
ax.plot(xvals, y_2_1, label=r"$\mu=2, \sigma=1$")
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("PDF($x$)", fontsize=20)
ax.tick_params(axis="both", labelsize=14)
leg = ax.legend(loc=0, prop={"size": 14})

fig.savefig("gaussian_multi_pdf.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.linspace(-4, 4, num=1000)
yvals = stats.norm.pdf(xvals, 0, 1)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, yvals, linewidth=3)

x1 = np.linspace(-1, 1, num=1000)
y1 = stats.norm.pdf(x1, 0, 1)
ax.fill_between(x1, y1, color="C1", alpha=0.9)
ax.text(0.7, 0.35, r"\textbf{68\%}", color="C1", fontsize=20)

x2 = np.linspace(-2, 2, num=1000)
y2 = stats.norm.pdf(x2, 0, 1)
ax.fill_between(x2, y2, color="C2", alpha=0.9, where=np.logical_or(x2 < -1, x2 > 1))
ax.text(1.5, 0.16, r"\textbf{95\%}", color="C2", fontsize=20)

x3 = np.linspace(-3, 3, num=1000)
y3 = stats.norm.pdf(x3, 0, 1)
ax.fill_between(x3, y3, color="C3", alpha=0.9, where=np.logical_or(x3 < -2, x3 > 2))
ax.text(2.5, 0.03, r"\textbf{99\%}", color="C3", fontsize=20)

ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("PDF($x$)", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("ci.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.arange(0, 40)
y1 = stats.binom.pmf(xvals, 20, 0.5)
y2 = stats.binom.pmf(xvals, 20, 0.75)
y3 = stats.binom.pmf(xvals, 40, 0.5)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals[xvals <= 20], y1[xvals <= 20], marker="o", linestyle="-", label=r"$n=20, p=0.5$")
ax.plot(xvals[xvals <= 20], y2[xvals <= 20], marker="s", linestyle="-", label=r"$n=20, p=0.8$")
ax.plot(xvals, y3, marker="d", linestyle="-", label=r"$n=40, p=0.5$")


ax.set_xlabel("Number of successes", fontsize=20)
ax.set_ylabel("Probability", fontsize=20)
ax.set_title("Binomial Distribution", fontsize=20)
ax.tick_params(axis="both", labelsize=14)
leg = ax.legend(loc=0, prop={"size": 14})

fig.savefig("binom.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.linspace(-3, 3, num=1000)
yvals = stats.uniform.pdf(xvals, -1, 3)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, yvals)
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("PDF($x$)", fontsize=20)
ax.set_title("$a = -1,\ b = 2$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("uniform_pdf.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.arange(0, 15)
y1 = stats.poisson.pmf(xvals, 1)
y2 = stats.poisson.pmf(xvals, 2.5)
y3 = stats.poisson.pmf(xvals, 10)

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, y1, marker="o", linestyle="-", label=r"$\lambda=1$")
ax.plot(xvals, y2, marker="s", linestyle="-", label=r"$\lambda=2.5$")
ax.plot(xvals, y3, marker="d", linestyle="-", label=r"$\lambda=10$")


ax.set_xlabel("Number of events", fontsize=20)
ax.set_ylabel("Probability", fontsize=20)
ax.set_title("Poisson Distribution", fontsize=20)
ax.tick_params(axis="both", labelsize=14)
leg = ax.legend(loc=0, prop={"size": 14})

fig.savefig("poisson.pdf", bbox_inches="tight", transparent=True)
```

```python
xvals = np.linspace(-5, 5, num=1000)
y1 = stats.norm.pdf(xvals, 0, 1)
y2 = stats.uniform.pdf(xvals, -np.sqrt(3), 2 * np.sqrt(3))

fig = plt.figure()
ax = plt.gca()
ax.plot(xvals, y1)
ax.plot(xvals, y2)

ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("$p(x)$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("compare_pdfs.pdf", bbox_inches="tight", transparent=True)
```
