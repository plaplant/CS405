```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc("text", usetex=True)
matplotlib.rc("font", family="serif")
matplotlib.style.use("dark_background")
```

```python
rng = np.random.default_rng()
xvals = rng.normal(loc=0.0, scale=3.0, size=(100,))
yvals = rng.normal(loc=3.2, scale=5.0, size=(100,))
yvals = yvals + 2 * xvals
```

```python
fig = plt.figure()
ax = plt.gca()
ax.scatter(xvals, yvals)
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("$y$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("scatter.pdf", bbox_inches="tight", transparent=True)
```

```python
fig = plt.figure()
ax = plt.gca()
ax.scatter(xvals, yvals + 2 * xvals**2)
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("$y$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("scatter2.pdf", bbox_inches="tight", transparent=True)
```

```python
tvals = rng.uniform(0, 2 * np.pi, size=(100,))
noise = rng.normal(loc=0.0, scale=0.5, size=(100,))
amp = 3.2
omega = 1.2
phi = 0.5
posvals = amp * np.cos(omega * tvals + phi) + noise
```

```python
fig = plt.figure()
ax = plt.gca()
ax.scatter(tvals, posvals)
ax.set_xlabel("$t$", fontsize=20)
ax.set_ylabel("$x(t)$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("sho.pdf", bbox_inches="tight", transparent=True)
```

```python
from scipy.stats import linregress

result = linregress(xvals, yvals)
```

```python
# plot the results of the regression
fig = plt.figure()
ax = plt.gca()
ax.scatter(xvals, yvals, color="C0")

xregress = np.linspace(-10, 10, num=1000)
yregress = result.slope * xregress + result.intercept
ax.plot(xregress, yregress, color="C1")

# also plot spread in slope values
ymin = (result.slope - result.stderr) * xregress + result.intercept
ymax = (result.slope + result.stderr) * xregress + result.intercept
ax.fill_between(xregress, ymin, ymax, color="C1", alpha=0.7)

# add labels
ax.set_xlabel("$x$", fontsize=20)
ax.set_ylabel("$y$", fontsize=20)
ax.tick_params(axis="both", labelsize=14)

fig.savefig("regress.pdf", bbox_inches="tight", transparent=True)
```
